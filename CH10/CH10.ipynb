## Loading and Preprocessing

```
import numpy as np
import pandas as pd

# load data
data = pd.read_csv('haberman.data', names=['Age at Operation','Year of Operation', 'Number of Detections','Survival'])
data
```

```
type(data)
```

```
data.drop(labels=['Year of Operation'], axis=1, inplace=True)
data
```

```
x = data[['Age at Operation', 'Number of Detections']]
y = data[['Survival']]
```

```
x
```

```
y
```

```
z = [1 if i==1 else 0 for i in y['Survival']]
z
```

## Simple model

```
# load libs
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
l_r = LogisticRegression(class_weight='balanced')
scalar = StandardScaler()
X = scalar.fit_transform(x)
model = l_r.fit(X,z)
```

```
model.predict([[56,9]]), model.predict([[55,1]])   
```

## Regularization

```
# to reduce variance
# load libs
from sklearn.linear_model import LogisticRegressionCV
l_rcv = LogisticRegressionCV(penalty='l2',Cs=10,n_jobs=-1)
model = l_rcv.fit(X,z)
```

```
model.predict([[56,9]]), model.predict([[55,1]])   
```
