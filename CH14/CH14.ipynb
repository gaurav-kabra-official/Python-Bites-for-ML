```
# load lib and image
import cv2
import matplotlib.pyplot as plt

image_plane = cv2.imread('plane.jpg')
plt.imshow(image_plane)
plt.show()
```

### resize & save

```
image_256 = cv2.resize(image_plane, (256,256))
cv2.imwrite('plane_256.jpg', image_256)
```
